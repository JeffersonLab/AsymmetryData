# Study of Proton Radius Extraction Methods

### Using A Realistic  Charge Form Factor To Understand Proton Radius Extractions

Here we are using a proton form factor G$_E$(q$^2$) function provided by Jose and Christian which has proper analytic properties.   As this is a constructed distribution, we know the "true" radius is 0.85fm allowing for a clear illistration of bias vs. variance.



```python
# Loading Ge Function

import pickle 

with open('interpolator_fm.pkl', 'rb') as f:
    Jose_fm = pickle.load(f)
    
# Reference Functions

def sd_fm(q2):
    return (1+0.11*q2/2)**(-2)

def dd_fm(q2):
    return (1+0.1176*q2/2)**(-2)
        
# Making Test Plot

import numpy as np
import matplotlib.pyplot as plt

q2_fm=np.linspace(0,10,1000)
plt.figure(dpi=150)
plt.plot(q2_fm,Jose_fm(q2_fm),'-',color='black',label='Jose and Christian Ge Function')
plt.plot(q2_fm,sd_fm(q2_fm),'-',color='blue',label='Standard 0.81fm Dipole')
plt.plot(q2_fm,dd_fm(q2_fm),'--',color='red',label='Doug 0.84fm Dipole')
plt.ylabel(r'G$_E$')
plt.xlabel(r'Q$^2$ [fm]$^{-2}$')
plt.xlim(0,10)
plt.ylim(0.3,1)
plt.legend()
plt.show()
```

# Example Generation Of Faux Data

Current experiments are limited to $Q^2$ down to 0.1 fm$^{-2}$ though this will soon be
extended by the PRAD experiment.

```python
# Generate Low Q2 Points With Random Error Sigma

np.random.seed(42)
q2_low = (np.sort(np.linspace(0.1,0.5,50)))
sigma=0.005

Jose_data=[]
for n in range(len(q2_low)):
    Jose_data.append(Jose_fm(q2_low[n])*(1+sigma*np.random.randn()))

# Plot The Result
plt.figure(dpi=150)
lines = {'linestyle': 'None'}
plt.rc('lines', **lines)
plt.errorbar(q2_low,Jose_data,fmt='',yerr=sigma,label='Faux Data')
plt.plot(q2_low,Jose_fm(q2_low),'-',color='black',label='True Function')
plt.ylabel(r'G$_E$')
plt.xlabel(r'Q$^2$ [fm]$^{-2}$')
plt.legend()
plt.show()
```

## Using LMFIT To Do Least Squares Regression

The LMFIT function makes use of the numpy least_squares and curve_fit functions while adding a number of nice features.   In particular, it has a nice way of added parameter bounds that is consistant with the way MINUIT (found in PAW and ROOT) solves bounded least square problems. 

```python
def linear_ge(q2,a,b):
    return a+b*q2

def quad_ge(q2,a,b,A):
    return a+b*q2+A*q2**2

def cubic_ge(q2,a,b,A,B):
    return a+b*q2+A*q2**2+B*q2**3

from lmfit import Model
from math  import sqrt

model=Model(quad_ge)
result=model.fit(Jose_data,q2=q2_low,a=1,b=-0.11,A=0.01,weights=1/sigma)

print(result.fit_report())
print('Extracted Raduis = {0:0.3f} fm'.format(sqrt(-6*result.best_values['b']/\
                                                result.best_values['a'])))

```

```python
plt.figure(dpi=150)
plt.plot(q2_low, Jose_fm(q2_low), '-', label='True Function')
plt.plot(q2_low, result.best_fit, 'r--', label='Best Fit')
plt.ylabel(r'G$_E$')
plt.xlabel(r'Q$^2$ [fm]$^{-2}$')
plt.legend()
plt.show()
```

## Now Make A Loop To Get Distributions Of Possible Results

```python
from lmfit import Parameters

loops=1000
radii=[]
radii_bound=[]
radii_linear=[]
radii_cubic=[]
radii_rational=[]

for m in range(loops):
    Jose_data=[]
    for n in range(len(q2_low)):
        Jose_data.append(Jose_fm(q2_low[n])*(1+sigma*np.random.randn()))
    #
    # Unbounded Least Squares Quadratic
    #
    model=Model(quad_ge)
    result=model.fit(Jose_data,q2=q2_low,a=1,b=-0.11,A=0.01,weights=1/sigma)
    radii.append(sqrt(-6*result.best_values['b']/result.best_values['a']))
    #
    model=Model(cubic_ge)
    result=model.fit(Jose_data,q2=q2_low,a=1,b=-0.11,A=0.01,B=0.01,weights=1/sigma)
    try:
        radii_cubic.append(sqrt(-6*result.best_values['b']/result.best_values['a']))
    except:
        print('Negative Sqrt')
    #
    # Now With Parameter Bounds
    #
    model=Model(quad_ge)
    params = Parameters()
    params.add('A', value=.0125)
    params.add('a', value=1)
    params.add('b', value=-0.1)
    params['A'].min=0.011 #0.011
    params['A'].max=0.015 #0.015
    result_bound=model.fit(Jose_data,q2=q2_low,params=params,weights=1/sigma)
    radii_bound.append(sqrt(-6*result_bound.best_values['b']/result_bound.best_values['a']))
    #
    # Linear Fit 
    #
    model=Model(linear_ge)
    result_linear=model.fit(Jose_data,q2=q2_low,a=1,b=-0.11,weights=1/sigma)
    radii_linear.append(sqrt(-6*result_linear.best_values['b']/result_linear.best_values['a']))
    #
    #
    # 
```

## Quadratic Fit

```python
from scipy.stats import norm
from matplotlib import mlab
(mu, sigma) = norm.fit(radii)

plt.figure(dpi=150)
n, bins, patches = plt.hist(radii, 60, range=[0.75, 0.95], normed=1, facecolor='gray', alpha=0.75)

# add a 'best fit' line
y = mlab.normpdf( bins, mu, sigma)
l = plt.plot(bins, y, 'b-', linewidth=2)

#plot
plt.xlabel('Radii [fm]')
plt.ylabel('Counts')
plt.title(r'$\mathrm{Quadratic\ Fit:}\ \mu=%.3f,\ \sigma=%.3f$' %(mu, sigma))
plt.grid(True)

plt.show()
```

```python
plt.figure(dpi=150)
(mu, sigma) = norm.fit(radii_cubic)
n, bins, patches = plt.hist(radii_cubic, 60, range=[0.75, 0.95], normed=1, facecolor='gray', alpha=0.75)

# add a 'best fit' line
y = mlab.normpdf( bins, mu, sigma)
l = plt.plot(bins, y, 'b-', linewidth=2)

#plot
plt.xlabel('Radii [fm]')
plt.ylabel('Counts')
plt.title(r'$\mathrm{Cubic\ Fit:}\ \mu=%.3f,\ \sigma=%.3f$' %(mu, sigma))
plt.grid(True)

plt.show()
```

## The Goldilocks Solution

The linear fit has a small variance with a bias and while the quadratic fit has no bias it has a huge variance.   The bounded quadratic fit finds the goldilocks solution with the low bias of a quadratic fit while also mantaining the low variance of the linear fit. 


## WARNING!!

This has all assumed that we have perfectly extracted form factors.   The hidden problem is the
division by the Mott cross section.    This division which is done when people repoeat reduced
cross sections hides a sensitively to angle offsets.  This problem was nicely noted by Dave
Mack as he reviewed the DRAD experiment.


## Rational Function Fit

Going to even higher order functions is a dangerous game as high order polynomials easily fit the random noise instead of the true signal, so finding a better function to use is important.
For this we try a low order rational function.   These functions are used in engineering for
extrapoling.

```python
# Generate High Q2 Points With Random Error Sigma

np.random.seed(42)
q2_low = (np.sort( np.linspace(0.1,10,500)))
sigma=0.003

Jose_data=[]
for n in range(len(q2_low)):
    Jose_data.append(Jose_fm(q2_low[n])*(1+sigma*np.random.randn()))

# Plot The Result
plt.figure(dpi=150)
lines = {'linestyle': 'None'}
plt.rc('lines', **lines)
plt.errorbar(q2_low,Jose_data,fmt='',yerr=sigma,label='Faux Data')
plt.plot(q2_low,Jose_fm(q2_low),'-',color='black',label="Jose and Christian's Function")
plt.ylabel(r'G$_E$')
plt.xlabel(r'Q$^2$ [fm]$^{-2}$')
plt.legend()
plt.show()
```

```python
from lmfit import Parameters

def rational_ge(q2,a,b,c):
    #
    return a/((1+b*q2)*(1+c*q2))

def cf_ge(q2,a,b,c):
    return a/(1+(b*q2/(1+(c*q2))))#/(1+d*q2))))) 

loops=1000

radii_rational=[]
radii_cf=[]

for m in range(loops):
    Jose_data=[]
    for n in range(len(q2_low)):
        Jose_data.append(Jose_fm(q2_low[n])*(1+sigma*np.random.randn()))
    # 
    # Rational Function
    #
    model=Model(rational_ge)
    result_rational=model.fit(Jose_data,q2=q2_low,a=1,b=-0.01,c=0.01,weights=1/sigma)
    radii_rational.append(sqrt(-6*(result_rational.best_values['b']
                                   -result_rational.best_values['c'])))
    #
    # Continued Fraction
    #
    model=Model(cf_ge)
    result_cf=model.fit(Jose_data,q2=q2_low,a=1,b=0.01,c=0.01,weights=1/sigma)
    radii_cf.append(sqrt(abs(6*(result_cf.best_values['b']))))
```

```python
print(result_rational.fit_report())
```

```python
print(result_cf.fit_report())
```

```python

(mu, sigma) = norm.fit(radii_rational)

plt.figure(dpi=150)
n, bins, patches = plt.hist(radii_rational, 480, range=[0.75,0.95], normed=1, facecolor='gray', alpha=0.75)

# add a 'best fit' line
y = mlab.normpdf( bins, mu, sigma)
l = plt.plot(bins, y, 'b-', linewidth=2)

#plot
plt.xlabel('Radii [fm]')
plt.ylabel('Counts')
plt.title(r'$\mathrm{Rational\ Function:}\ \mu=%.4f,\ \sigma=%.4f$' %(mu, sigma))
plt.grid(True)

plt.show()
```
